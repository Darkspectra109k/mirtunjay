```python
import matplotlib.pyplot as plt
```

```python
labels=['group1', 'group2', 'group3', 'group4']
men_ages=[20, 35, 30, 35]
women_ages=[25, 32, 34, 20]

fig, ax= plt.subplots()
ax.bar(labels,men_ages,label='men')
ax.bar(labels,women_ages,bottom=men_ages, label='women')

ax.set_ylabel('Ages')
ax.set_xlabel('Groups number')
ax.set_title('Ages by group')
ax.legend()
plt.show()
```

```python
import numpy as np
```

```python
classes=["python", "java", "c++", "c#"]
classs1_students=[20, 35, 30, 35]
classs2_students=[25, 32, 34, 20]

classes_index=np.arange(len(classes))
width=0.35

plt.bar(classes_index, classs1_students, width, label='class1_students')
plt.bar(classes_index + width, classs2_students, width, label='class2_students')
plt.xticks(classes_index + width , classes)
plt.ylabel('Number of students')
plt.xlabel('Classes')
plt.title('Number of students in classes')
plt.legend()
plt.show()  
```

```python
classes=["python", "java", "c++", "c#"]
classs1_students=[20, 35, 30, 35]
classs2_students=[25, 32, 34, 20]

classes_index=np.arange(len(classes))
width=0.35

plt.barh(classes_index, classs1_students, height=width, label='class1_students')
plt.barh(classes_index + width, classs2_students, height=width, label='class2_students')
plt.yticks(classes_index + width / 2, classes)
plt.ylabel('Number of students')
plt.xlabel('Classes')
plt.title('Number of students in classes')
plt.legend()
plt.show()  
```

create a dictoionary and list keys and values and assign them to respective variables

```python

my_dict = {'a': 1, 'b': 2, 'c': 3}

keys = list(my_dict.keys())
values = list(my_dict.values())
```

```python
plt.bar(keys, values)
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Bar Chart from Dictionary')
plt.show()
```

```python
plt.scatter(keys, values)
```

```python
x=100
plt.scatter(keys, values, s=x, c='red', alpha=0.5)
plt.show()
```

```python
x=400
plt.scatter(keys, values, s=x, marker='o')
plt.show()
```

```python
x_axis = np.array([0, 5, 10, 15, 20])
plt.plot(x_axis, x_axis**3180)
plt.legend(['x^3180'])
```

```python
y_axis = x_axis**2

empty_canvas = plt.figure()
create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b', label="x = x**2")
create_axis.plot(x_axis, y_axis + (x_axis * 3), 'g', label="x = y+(x*3)")

create_axis.legend(loc=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {'Batch 1': 10, 'Batch 2': 15, 'Batch 3': 5, 'Batch 4': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,3, figsize=(10, 3))

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

fig.suptitle('Subplots')
```
